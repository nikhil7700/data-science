```python
import numpy as np
```

```python
class solution:
    
    
    def get_emi(self,principal_amount,rate,tenure):
        monthly_rate = ((rate/12)/100)
        monthly_tenure = (tenure*12)
        num = principal_amount*monthly_rate*((1+monthly_rate)**monthly_tenure)
        den = (((1+monthly_rate)**monthly_tenure)-1)
        return round(num/den,2)
    
    
    def get_amount(Self,emi,rate,tenure):
        monthly_rate = ((rate/12)/100)
        monthly_tenure = tenure * 12
        num = emi*(((1+monthly_rate)**monthly_tenure)-1)
        den = monthly_rate*((1+monthly_rate)**monthly_tenure)
        return round(num/den,2)
    
    
    def get_interest_paid(self,principal_amount,rate,tenure,monthly_installment):
    
    
        monthly_rate = ((rate/12)/100)
        monthly_tenure = (tenure*12)
        count = 0
        interest_yearly_pay = 0
        principal_yearly_pay = 0
        principal_remaining = np.zeros(monthly_tenure+1)
        interest_yearly_pay_arr = np.zeros(tenure)
        principal_yearly_pay_arr = np.zeros(tenure)
        cumilative_interest_paid = 0
        month_to_complete_half_payment = 0
        flag = True


        for i in range(0,monthly_tenure+1):
            if i == 0:
                previous_principal_remaining = principal_amount
            else:
                previous_principal_remaining = principal_remaining[i-1]


            interest_payment = round(previous_principal_remaining*monthly_rate,2)
            principal_payment = round(monthly_installment-interest_payment,2)


            if previous_principal_remaining - principal_payment < 0:
                principal_payment = previous_principal_remaining


            interest_yearly_pay = interest_yearly_pay + interest_payment
            principal_yearly_pay = principal_yearly_pay + principal_payment

            principal_remaining[i] = previous_principal_remaining - principal_payment
            #print(interest_payment,principal_payment,principal_remaining[i],interest_payment+principal_payment)
            #print(principal_yearly_pay)
            if principal_yearly_pay >= (0.5*principal_amount) and flag:
                month_to_complete_half_payment = i+1
                flag = False

            if count%12==0 and count!=0:
                interest_yearly_pay_arr[(count//12)-1] = interest_yearly_pay
                principal_yearly_pay_arr[(count//12)-1] = principal_yearly_pay
                if count//12 == 1: 
                    print(round(interest_yearly_pay_arr[(count//12)-1],2))
                    cumilative_interest_paid = cumilative_interest_paid + interest_yearly_pay_arr[(count//12)-1]
                else:
                    print(round(interest_yearly_pay_arr[(count//12)-1] - interest_yearly_pay_arr[(count//12)-2],2))
                    cumilative_interest_paid = cumilative_interest_paid + (interest_yearly_pay_arr[(count//12)-1] - interest_yearly_pay_arr[(count//12)-2])
                #print(count)
            count = count + 1
        #for i in range(15):
         #   print(interest_yearly_pay_arr[i])



        return round(cumilative_interest_paid,2),month_to_complete_half_payment
     
```

```python
amount = float(input("Enter the Principal amount --> "))
rate = float(input("Enter the annual rate --> "))
flag = True
try:
    tenure = int(input("Enter the tenure--> "))
except Exception as e:
    print(e)
    print("tenure must be a integer ")
    flag = False
    
    
while True and flag:
    
    if amount<0 or rate < 0 or tenure < 0:
        print("amount or rate or tenure must be a positive value ")
        break
    else:
        obj = solution()
        EMI = obj.get_emi(amount,rate,tenure)
        print()
        print("The EMI of the loan amount {} at {} percent rate for {} years will be {}".format(amount,rate,tenure,EMI))
        new_EMI = float(input("Enter the new EMI--> "))
        new_rate = float(input("Enter the new annual rate-->"))
        if new_rate<0 or new_amount<0:
            print("rate or amount must be a positive value ")
            break
            
        affordable_Loan = obj.get_amount(new_amount,new_rate,tenure)
        print()
        print("The affordable Loan on {} EMI at {} percent rate for {} years will be {}".format(new_EMI,new_rate,tenure,affordable_Loan))
        print()
        print("The annual Interest paid for paid 15 years are as Follows")
        Cumilative_interest_paid,months = obj.get_interest_paid(amount,rate,tenure,EMI)
        print()
        print("The Cumilative Interest paid is {} for 15 years.".format(Cumilative_interest_paid))
        print()
        print("In {} months you hv completed 50% of the {} ".format(months,amount))
        break

    




        
    
    
    
```
