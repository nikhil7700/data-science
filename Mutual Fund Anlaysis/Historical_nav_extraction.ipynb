```python
import pandas as pd
```

```python
df = pd.read_pickle('historical_nav.pkl')
```

```python
df["data"]
```

```python


navs_data = {}
for index,row in df.iterrows():
    nav_data = {}
    for data in row["data"]:
        nav_data[data["date"]] = data["nav"]
        
    navs_data[row["scheme_code"]] = nav_data
#print(nav_data)
```

```python
print(navs_data)
```

```python
df2 = pd.DataFrame(navs_data)
df2.head()
```

```python
df2.to_pickle('historical_nav_final.pkl')
```
