# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing data set

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

# Training The Dataset in Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

# Training The dataset into Polynomial regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
x_poly = poly_reg.fit_transform(x)
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
```

```python
print(x_poly)
```

# Visualizing the result of simple Linear Regression

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

# Visualizing the plot in polynomial regression my solution(Wrong)

```python
plt.scatter(x,y,color='red')
plt.plot(x_poly,lin_reg2.predict(x_poly),color = 'blue')
plt.title('Position vs Salary(Linear Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

# Visualizing the Polynomial result with degree 2 with not smooth nor higher resolution

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg2.predict(poly_reg.fit_transform(x)),color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

# Training the Polynomial regression with degree 4 with more smooth and higher resolution

```python
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x)
lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,y)
```

# Visualizing the Polynomial result with degree 4 with more smooth and higher resolution

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg2.predict(poly_reg.fit_transform(x)),color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

# Further more smooth curve

```python
x_grid = np.arange(min(x),max(x),0.1)
x_grid = x_grid.reshape(len(x_grid),1)
plt.scatter(x,y,color='red')
plt.plot(x_grid,lin_reg2.predict(poly_reg.fit_transform(x_grid)),color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position')
plt.ylabel('Salary')
plt.show()
```

# Predicting Single Result using Linear Regression
